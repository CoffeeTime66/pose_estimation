```python
import cv2
import pandas as pd
import csv
import mediapipe as mp
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
```

```python
data = pd.read_csv("frames_pose.csv")
data
```

```python
mp_pose = mp.solutions.pose
```

```python
csv_file = open("pose200.csv", "w", newline='')
writer = csv.writer(csv_file)
header = []
header.append("Frame_path")
for i in range(17):
    header.append(f"X{i}")
    header.append(f"Y{i}")
    header.append(f"Z{i}")
header.append("label")
writer.writerow(header)


with mp_pose.Pose(
    static_image_mode=True,
    model_complexity=2,
    enable_segmentation=True,
    min_detection_confidence=0.5) as pose:
  for index, r in data.iterrows():
    poses = []
    image = cv2.imread(r['frame_path'])
    image_height, image_width, _ = image.shape
    # Convert the BGR image to RGB before processing.
    results = pose.process(image)

    if results.pose_landmarks:
        pose_landmarks = results.pose_landmarks
        poses.append(pose_landmarks)
    
    for i, pose_landmarks in enumerate(poses):
      row = []
      row.append(r['frame_path'])
      for landmark in pose_landmarks.landmark:
            row.append(landmark.x)
            row.append(landmark.y)
            row.append(landmark.z)
      row.append(r['label'])
      writer.writerow(row)        

csv_file.close()
```

```python
X = poses.iloc[:, :-1].values  # Ð˜ÑÐºÐ»ÑŽÑ‡Ð°ÐµÐ¼ ÑÑ‚Ð¾Ð»Ð±ÐµÑ† Ñ Ð½Ð¾Ð¼ÐµÑ€Ð°Ð¼Ð¸ ÐºÐ°Ð´Ñ€Ð¾Ð²
y = poses.iloc[:, -1].values
```

```python
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
```

```python
# ÐŸÑ€ÐµÐ¾Ð±Ñ€Ð°Ð·Ð¾Ð²Ð°Ð½Ð¸Ðµ Ð´Ð°Ð½Ð½Ñ‹Ñ… Ð² Ñ‚Ñ€ÐµÐ±ÑƒÐµÐ¼Ñ‹Ð¹ Ñ„Ð¾Ñ€Ð¼Ð°Ñ‚ Ð´Ð»Ñ Ð¼Ð¾Ð´ÐµÐ»Ð¸
# Ð¡Ð¾Ð·Ð´Ð°Ð½Ð¸Ðµ ÑÐ¿Ð¸ÑÐºÐ° Ð´Ð»Ñ Ñ…Ñ€Ð°Ð½ÐµÐ½Ð¸Ñ Ð¿Ð¾ÑÐ»ÐµÐ´Ð¾Ð²Ð°Ñ‚ÐµÐ»ÑŒÐ½Ð¾ÑÑ‚ÐµÐ¹ Ð¿Ð¾Ð·
sequences = []
labels = []
```

```python
sequence_length = 200  # Ð”Ð»Ð¸Ð½Ð° ÐºÐ°Ð¶Ð´Ð¾Ð¹ Ð¿Ð¾ÑÐ»ÐµÐ´Ð¾Ð²Ð°Ñ‚ÐµÐ»ÑŒÐ½Ð¾ÑÑ‚Ð¸
for i in range(0, len(X) - sequence_length + 1, sequence_length):
    sequence = X[i:i+sequence_length]
    label = y[i+sequence_length-1]  # ÐœÐµÑ‚ÐºÐ° Ð´Ð»Ñ Ð¿Ð¾ÑÐ»ÐµÐ´Ð½ÐµÐ³Ð¾ Ñ„Ñ€ÐµÐ¹Ð¼Ð° Ð² Ð¿Ð¾ÑÐ»ÐµÐ´Ð¾Ð²Ð°Ñ‚ÐµÐ»ÑŒÐ½Ð¾ÑÑ‚Ð¸
    sequences.append(sequence)
    labels.append(label)
```

```python
X = np.array(sequences)
y = np.array(labels)
```

```python
sequences[2]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2024)
```

```python
model = Sequential()
model.add(LSTM(128, input_shape=(X.shape[1], X.shape[2])))
model.add(Dropout(0.5))
model.add(Dense(64, activation="relu"))
model.add(Dense(len(label_encoder.classes_), activation="softmax"))
```

```python
model.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
model.fit(X_train, y_train, epochs=20, batch_size=32)
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Test Loss: {loss}, Test Accuracy: {accuracy}')
```
